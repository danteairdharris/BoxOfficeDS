```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
import pickle
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)

# create data frame for model relevant columns
df = pd.read_csv('./WB_movie_data_dropna.csv')
# display(df)
df_model = df[['Release_month', 'Running_time_min', 'Budget', 'Box_office', 'Box_office_opening', 'Rating']]

#get dummy variables for movie rating
df_dum = pd.get_dummies(df_model)
display(df_dum)
```

```python
x = df_dum.drop('Box_office_opening', axis=1)
y = df_dum.Box_office_opening.values

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

lm = LinearRegression()
lm.fit(X_train, y_train)

## Cross Validation
# np.mean(cross_val_score(lm, X_train, y_train, scoring='neg_mean_absolute_error', cv=3))

# predict box office opening among the x test set
predict = lm.predict(X_test)
# evaluate prediction across the y test set
mean_absolute_error(y_test, predict)
```

```python
## Save list_movie_data in with pickle
def save_data(filename, data):
    with open(filename, 'wb') as f:
        pickle.dump(data, f)
        
        
save_data('LinearRegressionModel.pickle', lm)
```
